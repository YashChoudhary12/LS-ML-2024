# Pandas Implementation

```python
import pandas as pd

# Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df = pd.DataFrame(data)

# Display the first 2 rows of the data frame
print("First 2 rows of the DataFrame:",df.head(2))

# Print the age column
print("\nAge column:")
print(df['Age'])

# Filter rows where age is greater than 26
filtered_df = df[df['Age'] > 26]
print("\nRows where age is greater than 26:")
print(filtered_df)

# Add a new column 'Country' with the value 'India' for all rows
df['Country'] = 'India'
print("\nDataFrame with new 'Country' column:")
print(df)

# Create a new DataFrame with missing values
data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df2 = pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
mean_age = df2['Age'].mean()
df2['Age'].fillna(mean_age, inplace=True)
print("\nDataFrame with missing values filled in 'Age' column:")
print(df2)
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the points
xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.figure()
plt.plot(xpoints, ypoints, 'o')  # 'o' indicates points without lines
plt.title("Points without line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Plotting with a star marker
plt.figure()
plt.plot(xpoints, ypoints, '*')  # '*' indicates star markers
plt.title("Points with Star Marker")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Using fmt format: add circular marker, red color and dashed line
plt.figure()
plt.plot(xpoints, ypoints, 'ro--')  # 'ro--' indicates red color, circular marker, and dashed line
plt.title("Circular Marker, Red Color, Dashed Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.figure()
plt.scatter(xpoints, ypoints)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Set color to the scatter plot: Blue, Green, Red, and Yellow for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.figure()
plt.scatter(xpoints, ypoints, c=colors)
plt.title("Scatter Plot with Different Colors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Miscellaneous Modules Implementation

```python
import random
import numpy as np

# Set the seed of random to 20
random.seed(20)

arr1 = np.array([1, 24, 31, 45, 73, 81, 94, 25])

# Using the random module, pick 4 different random numbers from arr1 and return their sum
random_numbers = random.sample(list(arr1), 4)
sum_of_random_numbers = sum(random_numbers)

print("Randomly selected numbers:", random_numbers)
print("Sum of randomly selected numbers:", sum_of_random_numbers)
```
